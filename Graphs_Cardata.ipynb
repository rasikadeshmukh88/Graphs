```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the car_data dataset
df= pd.read_csv("car_data.csv")
```

```python
df
```

```python
df.columns
```

# Bar Chart example

```python
plt.bar(df['Year'],df['Selling_Price'])
plt.xlabel('Years')
plt.ylabel('Selling price')
plt.title('Selling Price In Years')
plt.show()
```

```python
#Change the color using (color='color_name')
#Set the width of bars using (width=0.5)
plt.bar(df['Year'],df['Selling_Price'],color='violet',width=0.5)
plt.xlabel('Years')
plt.ylabel('Selling price')
plt.title('Selling Price In Years')
plt.show()
```

### For horizontal bar plots, you can use the barh() function.

```python
plt.barh(df['Year'],df['Selling_Price'])
plt.xlabel('Selling Price')
plt.ylabel('Years')
plt.title('Selling Price In Years')
plt.show()
```

# HeatMap Example

```python
# selected the numeric column
selected_columns = ['Year', 'Present_Price', 'Selling_Price']

coor = df[selected_columns].corr() 
sns.heatmap(coor,annot = True, cmap='YlGnBu')
plt.title("Correlation Heatmap (Selected Columns)")
plt.show()
```

```python
plt.hist(['Year'])
plt.show()
```

```python
sns.pairplot(df, hue = "Year")
plt.show()
```

```python
sns.boxplot(x = "Year", y= "Present_Price", data = df)
plt.title("Box Plot")
plt.show()
```

```python
sns.histplot(df['Present_Price'], kde = True)  # kernel density estimation
plt.title("Histplot of Present_Price")
plt.show()
```

```python
# Bar Plot
sns.barplot(x = "Year", y ="Selling_Price", data = df)
plt.title("Bar Plot day vs total_bill")
plt.show()
```

```python
# Line Plot
sns.lineplot(x ="Selling_Price", y = "Year", data = df)
plt.title("Total bill vs Tips")
plt.grid()
plt.show()
```

```python
# Line Plot
sns.lineplot(x ="Present_Price", y = "Year", data = df)
plt.title("Line Plot")
plt.grid()
plt.show()
```
